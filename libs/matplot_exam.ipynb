```python
import matplotlib
import matplotlib.font_manager as fm

fig = plt.figure(constrained_layout = True)

#원하는 폰트 파일 경로 문자열로 입력
font_location = ""
#지정한 폰트 파일 경로에서 폰트 추출
fontname = fm.FontProperties(fname=font_location).get_name()

#matplotlib의 전역 폰트 설정 변경
matplotlib.rc('font',family = fontname)

a1 = fig.add_subplot(4,1,4)

fig = plt.figure()
#행개수,열개수,위치값
fig.add_subplot(4,1,4)
fig.add_subplot(2,4,2)

x = range(0,10)
y = range(5,15)


plt.show()
```

```python
fig = plt.figure()

plt.style.use("ggplot")

a = fig.add_subplot(111)
x = [3,4]
y = [1,2]

labels = ['english', 'math']
colors = ['red','blue']

a.barh(x,y, tick_label = labels, color = colors)

for i, v in enumerate(y):
    a.test(v+10, i+1.0, str(v))

plt.show()
```
